## 2. 为什么预训练不够

预训练学到 next-token prediction，但用户需要遵循指令、拒绝不当请求、稳定格式和有用回答。


## 3. Instruction data

SFT 数据通常是 prompt-response 对，直接教模型模仿目标行为。


## 4. Preference data

偏好数据是同一 prompt 下多个回答的相对好坏，用于 reward model、DPO 或 RLHF。


## 5. SFT 与偏好优化

SFT 学会格式和基础行为；偏好优化进一步推动更符合人类或规则偏好的输出。


## 6. 规模问题

post-training 数据更贵、更少，但对模型交互行为影响很大。

## 动手实验 2：DPO loss 形状玩具版

先直接运行，再改一个数字或字符串。


```python
import math

def log_sigmoid(x):
    return -math.log1p(math.exp(-x))

beta = 0.1
chosen_delta = 3.0 - 2.0
rejected_delta = 1.0 - 1.5
dpo_loss = -log_sigmoid(beta * (chosen_delta - rejected_delta))
print("DPO loss:", dpo_loss)
```
