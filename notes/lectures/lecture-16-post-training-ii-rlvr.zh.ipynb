## 2. 为什么可验证奖励重要

如果答案能自动验证，就不完全依赖人类偏好模型，RL 信号更清楚也更容易扩展。


## 3. PPO / GRPO 直觉

策略生成多个回答，根据 reward 调整概率，同时限制模型不要偏离太远。


## 4. Long-CoT

RL 可能鼓励模型生成更长的推理轨迹，只要这些轨迹提高可验证任务得分。


## 5. SFT vs RL

SFT 模仿已有答案；RL 可以探索新策略，但更不稳定。


## 6. 系统问题

RL 需要大量采样、打分和训练循环，inference 成本也成为训练成本的一部分。


## 动手实验 1：Group-relative advantage 玩具版

先直接运行，再改一个数字或字符串。


```python
rewards = [0, 1, 1, 0, 2]
mean = sum(rewards) / len(rewards)
advantages = [r - mean for r in rewards]
print("rewards:", rewards)
print("mean:", mean)
print("advantages:", advantages)
```

## 动手实验 2：带 KL 惩罚的 reward

先直接运行，再改一个数字或字符串。


```python
task_reward = 1.0
kl_to_reference = 0.3
beta = 0.05
objective = task_reward - beta * kl_to_reference
print("objective:", objective)
```
